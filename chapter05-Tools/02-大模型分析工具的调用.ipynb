## 举例2：如何调用具体大模型分析出来的工具
旧版本举例：

```python
from langchain_community.tools import MoveFileTool
from langchain_core.messages import HumanMessage
from langchain_core.utils.function_calling import convert_to_openai_function
from langchain_openai import ChatOpenAI
import os
import dotenv

dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL"),
    temperature=0,
    model="gpt-4o-mini",
)

#2.获取工具列表
tools=[MoveFileTool()]

#3.
functions=[convert_to_openai_function(t) for t in tools]

#4.获取消息列表
messages=[HumanMessage(content="将文件D:\\桌面\\实践报告\\大模型生成材料结构实践报告.docx移动到桌面上")]

response=llm.invoke(
    input=messages,
    functions=functions,
)
print(response)
```

步骤一：调用工具、函数分析

```python
import json
if "function_call" in response.additional_kwargs:
    tool_name=response.additional_kwargs["function_call"]["name"]
    #response.additional_kwargs["function_call"]["arguments"]拿到的不是python字典，而是json字符串，需要转换
    tools_args=json.loads(response.additional_kwargs["function_call"]["arguments"]) #<class 'dict'>
    # tools_args=response.additional_kwargs["function_call"]["arguments"]  #<class 'str'>
    print(f"调用工具：{tool_name} \n参数：{tools_args}")
    print(type(tool_name))
    print(type(tools_args))
else:
    print(f"没有调用工具,模型回复：{response.content}")
```

步骤二：调用对应的工具

```python
if "move_file" in response.additional_kwargs["function_call"]["name"]:
    tool=MoveFileTool()
    result=tool.run(tools_args)
    print(f"工具返回结果：{result}")
```

# 新版举例：

```python
from langchain_community.tools import MoveFileTool
from langchain_core.messages import HumanMessage
from langchain_openai import ChatOpenAI
import os
import dotenv

dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    base_url=os.getenv("OPENAI_BASE_URL"),
    temperature=0,
    model="gpt-4o-mini",
)


tools=[MoveFileTool()]

#3.返回已经绑定了工具的聊天模型 Runnable
llm_with_tools=llm.bind_tools(tools)

#4.获取消息列表
messages=[HumanMessage(content="将文件D:\\桌面\\实践报告\\大模型生成材料结构实践报告.docx移动到桌面上")]
response=llm_with_tools.invoke(messages)

print(response)
```

```python
if response.tool_calls:
    tool_name=response.tool_calls[0]["name"]
    tools_args=response.tool_calls[0]["args"]
else :
    print(f"没有调用工具,模型回复：{response.content}")
print(f"调用工具：{tool_name} \n参数：{tools_args}")
print(type(tool_name))
print(type(tools_args))
```

```python
if "move_file" in tool_name:
    tool=MoveFileTool()
    result=tool.run(tools_args)
    print(f"工具返回结果：{result}")
```

## create_agent
### 这里的 graph 不是普通模型，而是一个 LangGraph 风格的 agent。
### agent 运行时处理的不是一个单独的 message list，而是一个 state（状态）对象
create_agent 支持直接传 Python 函数作为工具

```python
from langchain.agents import create_agent

def check_weather(location: str) -> str:
    '''Return the weather forecast for the specified location.'''
    return f"It's always sunny in {location}"

#这里的 graph 不是普通模型，而是一个 LangGraph 风格的 agent。
#agent 运行时处理的不是一个单独的 message list，而是一个 state（状态）对象
graph = create_agent(
    model=llm,
    tools=[check_weather],
    system_prompt="You are a helpful assistant",
)

#####传状态字典，不是消息列表了#######
inputs = {"messages": [{"role": "user", "content": "what is the weather in sf"}]}
for chunk in graph.stream(inputs, stream_mode="updates"):
    print(chunk)
```

## create_react_agent
### 进入弃用期

```python
from langgraph.prebuilt import create_react_agent

def check_weather(location: str) -> str:
    """Return the weather forecast for the specified location."""
    return f"It's always sunny in {location}"

graph = create_react_agent(
    model=llm,
    tools=[check_weather],
    prompt="You are a helpful assistant",
)

inputs = {
    "messages": [
        {"role": "user", "content": "what is the weather in sf"}
    ]
}

for chunk in graph.stream(inputs, stream_mode="updates"):
    print(chunk)
```
