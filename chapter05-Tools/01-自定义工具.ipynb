## 使用@tool 装饰器定义工具

```python
from langchain.tools import tool
from langchain_core.tools import StructuredTool


@tool(name_or_callable="add_two_number", description="add two numbers", return_direct=True)
def add_number(a:int,b:int)->int:
    """计算两个整数的和"""
    return a+b

print(f"name={add_number.name}")  #默认是函数的名称
print(f"args={add_number.args}")
print(f"description={add_number.description}")  ## 默认是函数的描述信息
print(f"return_direct={add_number.return_direct}")  ##默认值是false，仅对agent有效，为True时在调用给定工具后，agent将停止并将结果直接返回给用户
```

## 调用工具

```python
add_number.invoke({"a":100, "b":200})
```

## 修改args参数的描述

```python
from pydantic import BaseModel, Field
from langchain.tools import tool
#定义参数描述
class FieldInfo(BaseModel):
    a: int = Field(description="第一个整型参数")#'description': '第一个整型参数'
    b: int = Field(description="第二个整型参数")#'description': '第二个整型参数'

##加上 @tool(...) 之后，它就不再只是一个普通函数了，而是被 LangChain 改造成了一个工具对象，
# 这样 agent 或大模型就能识别和调用它。
@tool(name_or_callable="add_two_number", description="add two numbers", return_direct=True, args_schema=FieldInfo)
def add_number(a:int,b:int)->int:
    """计算两个整数的和"""
    return a+b

print(f"name={add_number.name}")  #默认是函数的名称
print(f"args={add_number.args}")
print(f"description={add_number.description}")  ## 默认是函数的描述信息
print(f"return_direct={add_number.return_direct}")  ##默认值是false，
#return_direct=True：
#如果 agent 调用了这个工具，那么工具返回结果后，agent 可以直接把结果返回给用户，不再继续后续推理流程。
```

## StructuredTool的from_function()方法的使用

```python
from langchain_core.tools import StructuredTool

#声明一个函数
def search_google(query:str):
    return "查询结果"

#定义一个工具对象
search_01=StructuredTool.from_function(
    func=search_google,
    name="search_google",
    description="google搜索",
)

print(f"name={search_01.name}")
print(f"args={search_01.args}")
print(f"description={search_01.description}")
```

```python
search_01.invoke({"query":"langchain"})
```

### 举例2：

```python
from langchain_core.tools import StructuredTool,FieldInfo


#定义参数描述
class FieldInfo(BaseModel):
    query: str = Field(description="查询参数")#'description': '查询参数'


#声明一个函数
def search_google(query:str):
    return "查询结果"

search_02=StructuredTool.from_function(
    func=search_google,
    name="search_google",
    description="google搜索",
    return_direct=True,
    args_schema=FieldInfo,
)

print(f"name={search_02.name}")
print(f"args={search_02.args}")
print(f"description={search_02.description}")
print(f"return_direct={search_02.return_direct}")
```
