1、获取大模型

```python
from langchain_openai import ChatOpenAI
import langchain
import os
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url="https://api.deepseek.com",
    temperature=0,
    model="deepseek-chat",
)
response=llm.invoke("介绍一下你自己")
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate
import os
import dotenv

dotenv.load_dotenv()

llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0.7,
    model="deepseek-chat",
)

#需要注意的一点是，这里需要指明具体的role，比如system、user
prompt=ChatPromptTemplate.from_messages( [
    ("system", "你是世界级的技术文档编写者"),
    ("user", "{input}"),
])
print(type(prompt))
chain=prompt | llm
response=chain.invoke({"input":"写一首情诗"})
print(response.content)
```

### 调用方式2：

```python
from langchain_core.messages import HumanMessage, SystemMessage
import os
import dotenv

dotenv.load_dotenv()

os.environ["DEEPSEEK_API_KEY"]=os.getenv("DEEPSEEK_API_KEY")
os.environ["DEEPSEEK_BASE_URL"]=os.getenv("DEEPSEEK_BASE_URL")
llm=ChatOpenAI(
    temperature=0.7,
    model="deepseek-chat",
)
prompt_template=ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="你是一位AI大师"),
        HumanMessage(content="你好")
    ]
)
print(type(prompt_template))     #<class 'langchain_core.prompts.chat.ChatPromptTemplate'>
prompt=prompt_template.invoke({})
print(type(prompt))              #<class 'langchain_core.prompt_values.ChatPromptValue'>
response=llm.invoke(prompt)
print(response)
```

### llm.invoke() 要的不是 ChatPromptTemplate而是这几类之一：

PromptValue

str

list[BaseMessage]
