### 1、PromptTemplate如何获取实例
### 方式一：使用构造方法的方式

```python
from langchain_core.prompts import PromptTemplate

prompt_template=PromptTemplate(
    input_variables=["name","age"],
    template="{name} is a good teacher,the age is {age}"
)
prompt=prompt_template.format(name="小丽",age="22")
print(prompt)
```

### 方式2：使用from_template()方法,如果提示词模版中没有变量，也可以使用from_template()方法

```python
prompt_template=PromptTemplate.from_template(
    template="{name} is a good teacher,the age is {age}"
)
prompt=prompt_template.format(name="小丽",age="22")
print(prompt)
```

## 2、部分提示词模版的使用（重点）

```python
prompt_template=PromptTemplate.from_template(
    template="{name} is a good teacher,the age is {age}",
    partial_variables={"name":"小丽"}
)
prompt=prompt_template.format(age="22")
print(prompt)
```

### 方式2.partial()调用

```python
from langchain_core.prompts import PromptTemplate

prompt_template=PromptTemplate(
    input_variables=["name","age"],
    template="{name} is a good teacher,the age is {age}"
)
#partial()调用完以后不会对调用者这个模版产生影响；而其返回值是一个新的模版
template1=prompt_template.partial(name="小丽")
prompt=template1.format(age="22")
print(prompt)
```

```python
prompt_template=PromptTemplate(
    input_variables=["name","age"],
    template="{name} is a good teacher,the age is {age}"
).partial(name="小丽")
prompt=prompt_template.format(age="22")
print(prompt)
```

### 给变量赋值的两种方法：format()/invoke()
#### format():参数部分：给变量赋值；返回值：str类型
#### invoke():参数部分：使用的是字典；返回值：PromptValue类型
##### 举例1：format()方法

```python
prompt_template=PromptTemplate(
    input_variables=["name","age"],
    template="{name} is a good teacher,the age is {age}"
).partial(name="小丽")
prompt=prompt_template.format(age="22")
print(type(prompt))
```

##### 举例2：invoke()方法  返回的PromptValue 可以传给LLM或者ChatModel

```python
prompt_template=PromptTemplate(
    input_variables=["name","age"],
    template="{name} is a good teacher,the age is {age}"
)
prompt=prompt_template.invoke({"age":"22","name":"小丽"})
print(type(prompt))#StringPromptValue
```

## 结合大模型使用

```python
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_openai import ChatOpenAI
import os
import dotenv
dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)

## 生成提示词模版
prompt_template=PromptTemplate.from_template(
    template="{name} is a good teacher,the age is {age}",
)
## 模版变量赋值
prompt=prompt_template.invoke(input={"age":"22","name":"小丽"})
## 调用大模型
response=llm.invoke(prompt)
print(response.content)
print(type(response))##<class 'langchain_core.messages.ai.AIMessage'>
```
