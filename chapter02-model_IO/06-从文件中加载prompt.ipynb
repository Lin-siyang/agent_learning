# 从物理磁盘的yaml中读取prompt

```python
import dotenv
from langchain_core.prompts import load_prompt

dotenv.load_dotenv()

prompt=load_prompt("prompt.yaml", encoding="utf-8")
print(prompt.format(name="年轻人", what="langchain"))
```
