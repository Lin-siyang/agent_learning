# 关于对话模型中消息（message）的使用
## 参考LangChain API reference文档

```python
import time

from langchain_core.messages import SystemMessage, HumanMessage
from langchain_openai import ChatOpenAI
import os
import dotenv
dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)
systemMessage=SystemMessage(content="你是一位经验丰富的老师")
humanMessage=HumanMessage(content="你好")
messages=[systemMessage,humanMessage]
response=llm.invoke(messages)
print(response.content)
```

# 多轮对话与上下文记忆

```python
sys_message=SystemMessage(content="你是一个助手,名字叫小丽")
human_message=HumanMessage(content="你喜欢什么书")
human_message1=HumanMessage(content="你叫什么名字")

messages=[sys_message,human_message,human_message1]
response=llm.invoke(messages)
print(response.content)
```

# 关于模型调用的方法说明
## invoke()/stream()
## batch():批量调用
## ainvoke()/astream()/abatch()：异步调用
### 举例1：stream()

```python
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0.7,
    model="deepseek-chat",
    streaming=True,#启用流式输出，默认是阻塞式输出false
)
messages=[HumanMessage(content="请介绍一下你自己")]
print("开始流式输出...")
for chunk in llm.stream(messages):
    print(chunk.content,end="",flush=True)#end：不换行，flush：强制刷新缓冲区
print("\n结束流式输出...")
```

### 举例2：batch()批量调用

```python
Message1=[SystemMessage(content="你是一个助手"),
          HumanMessage(content="三加一等于几？")
          ]
Message2=[SystemMessage(content="你一个助手"),
          HumanMessage(content="三乘五等于几？")
          ]
Message3=[SystemMessage(content="你一个助手"),
          HumanMessage(content="三除以五等于几？")
          ]
messages=[Message1,Message2,Message3]
response=llm.batch(messages)
print(response)#批量调用，返回一个列表，有三个AIMessage对象
```

### 同步异步方法调用

```python
import time
def call_model():
    print("开始调用模型...")
    time.sleep(5)
    print("结束调用模型...")

def perform_task():

    for i in range(3):
        print(f"开始执行任务{i+1}...")
        time.sleep(1)

def main():
    start_time=time.time()
    call_model()
    perform_task()
    end_time=time.time()
    return end_time-start_time

main_time=main()
print(f"总共耗时{main_time}秒")#阻塞式，同步
```

### 异步

```python
import asyncio

async def async_call(llm):
    await asyncio.sleep(5)
    print("结束调用模型...")

async def performer_other_task():
    await asyncio.sleep(5)
    print("结束其他任务...")

async def main():
    start_time=time.time()
    await asyncio.gather(async_call(None),performer_other_task())
    end_time=time.time()
    return f"总共耗时{end_time-start_time}秒"
#在jupyter notebook中运行的方式
result=await main()
print(result)


# #正确运行异步任务的方式
# if  __name__ == "__main__":
#     #使用asyncio.run()来启用异步任务
#     result=asyncio.run(main())
#     print(result)
```
