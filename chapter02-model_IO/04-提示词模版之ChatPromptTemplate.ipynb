#### 1、实例化的方式
#### 方式一：使用构造方法

```python
from langchain_core.prompts import ChatPromptTemplate

from hello_qwen import response

chat_prompt_template=ChatPromptTemplate(
    [
        ("system","你是一个AI助手，你的名字是{name}"),
        ("human","我的问题是{question}")
    ]
)
response=chat_prompt_template.invoke({"name":"小丽","question":"你可以干什么？"})
print(response)
##messages=[SystemMessage(content='你是一个AI助手，你的名字是小丽', additional_kwargs={}, response_metadata={}), HumanMessage(content='我的问题是你可以干什么？', additional_kwargs={}, response_metadata={})]
# 列表里面有两个消息
print(type(response))
#<class 'langchain_core.prompt_values.ChatPromptValue'>
```

##### 方式2：from_messages()

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字是{name}"),
    ("human","我的问题是{question}")
])
response=chat_prompt_template.invoke({"name":"小丽","question":"你可以干什么？"})
print(response)
print(type(response))
```

### 2、调用提示词模版的几种情况
#### 1、invoke() 传入的是字典，返回ChatPromptValue

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字是{name}"),
    ("human","我的问题是{question}")
])
response=chat_prompt_template.invoke({"name":"小丽","question":"你可以干什么？"})
print(response)
print(type(response))
```

#### format()传入变量的值，返回str

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字是{name}"),
    ("human","我的问题是{question}")
])
response=chat_prompt_template.format(name="小丽",question="你可以干什么？")
print(response)
print(type(response))#<class 'str'>
```

#### format_messages()传入变量的值，返回list

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字是{name}"),
    ("human","我的问题是{question}")
])
response=chat_prompt_template.format_messages(name="小丽",question="你可以干什么？")
print(response)
print(type(response))#<class 'list'>
```

#### format_prompt()传入变量的值，返回ChatPromptValue

```python
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字是{name}"),
    ("human","我的问题是{question}")
])
response=chat_prompt_template.format_prompt(name="小丽",question="你可以干什么？")
print(response)
print(type(response))#<class 'langchain_core.prompt_values.ChatPromptValue'>
```

### 实现ChatPromptTemplate与字符串和列表的转换
#### format_prompt()传入变量的值，返回ChatPromptValue
#### format_messages()传入变量的值，返回list
#### format()传入变量的值，返回str
#### invoke() 传入的是字典，返回ChatPromptValue

```python
#将ChatPromptValue转换成消息构成的list列表
response_messages=response.to_messages()
print(response_messages)
print(type(response_messages))

#将ChatPromptValue转换成字符串
response_str=response.to_string()
print(response_str)
print(type(response_str))
```

 ## 3、更丰富的实例化参数类型
#### 本质：不管是使用实例化的方式还是from_message()的方式来创建ChatPromptTemplate，传入的都是消息列表
#### 参数messages是一个列表，而列表里的元素的类型是多样的，可以是字符串，也可以是字典，消息类型，元祖够成的列表

```python
from langchain_core.prompts import ChatPromptTemplate
# 字符串str
chat_prompt_template1=ChatPromptTemplate.from_messages(
    [
        "你的名字是{name}"#默认是HumanMessage
    ]
)
response1=chat_prompt_template1.invoke({"name":"小丽"})
print(response1)
print(type(response1))

#字典类型
chat_prompt_template2=ChatPromptTemplate.from_messages(
    [
    {"role":"system","content":"你的名字是{name}"},
    {"role":"human","content":"我的问题是{question}"}
    ]
)
response2=chat_prompt_template2.invoke({"name":"小丽","question":"你可以干什么？"})
print(response2)
print(type(response2))

```

## 结合LLM
### 1、提供大模型
### 2、通过Chat提示词模版创建提示词
### 3、大模型调用提示词

```python
from langchain_openai import ChatOpenAI
import os
import dotenv
dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)
#创建提示词
chat_prompt_template=ChatPromptTemplate.from_messages([
    ("system","你是一个AI助手，你的名字是{name}"),
    ("human","我的问题是{question}")
])
prompt=chat_prompt_template.invoke({"name":"小丽","question":"你可以干什么？"})
response=llm.invoke(prompt)
print(response)
```

## 5、插入消息列表MessagesPlaceholder

```python
from langchain_core.messages import AIMessage, SystemMessage, HumanMessage
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="You are a helpful assistant that answers in pirate language"),
        MessagesPlaceholder("history"),
        ("human", "{question}"),
    ]
)
response=prompt.invoke(
    {
        "history": [("human", "what's 5 + 2"), AIMessage(content="seven")],
        "question": "now multiply that by 4",
    }
)
response1=prompt.format_messages(
    question="what's 5 + 2",
    history=[HumanMessage(content="what's 5 + 2"), AIMessage(content="seven")]
)
print(response)
print(type(response))
print(response1)
print(type(response1))
```
