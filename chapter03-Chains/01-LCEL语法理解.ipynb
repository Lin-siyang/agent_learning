### 使用管道符 | 之前的举例：

```python
from langchain_classic.chains.llm import LLMChain
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser, JsonOutputParser
from langchain_openai import ChatOpenAI
import os
import dotenv


dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)

chat_prompt_template=ChatPromptTemplate.from_messages(
    [
        ("system","你是一个AI助手，名字叫{role}"),
        ("human","{question}"),
        ("human","回答的格式为{format}")
    ]
)
parser=JsonOutputParser()
parser_out=parser.get_format_instructions()
prompt=chat_prompt_template.format(role="小丽",question="你叫什么名字",format=parser_out)
print(llm.invoke(prompt).content)
```

### 使用 | 举例：

```python
chain=chat_prompt_template | llm | parser
print(chain.invoke({"role":"小丽","question":"你叫什么名字","format":parser_out}))
```

## 传统Chain的使用：LLMChain ：弃用的类

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
import os
import dotenv
from langchain_classic.chains import LLMChain


dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)
prompt_template=PromptTemplate.from_template(
    template="帮我解决如下的问题：{question}",
)
chain = LLMChain(
    llm=llm,
    prompt=prompt_template,
    verbose=True,##显示执行过程中的详细日志情况
)

response=chain.invoke(input={"question":"1+2=?"})
print(response)
```
