### SimpleSequentialChain的使用
### 每个模块只有一个输入和一个输出

```python
from langchain_classic.chains.sequential import SimpleSequentialChain
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_openai import ChatOpenAI
import os
import dotenv
from langchain_classic.chains import LLMChain


dotenv.load_dotenv()
llm=ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)
chianA_template=ChatPromptTemplate.from_messages(
    [
        ("system","你是一个AI助手"),
        ("human","我的问题是{question}"),
    ]
)
chainA_chains = LLMChain(
    llm=llm,
    prompt=chianA_template,
    verbose=True,##显示执行过程中的详细日志情况
)
```

```python
chainB_template=ChatPromptTemplate.from_messages(
    [
        ("system","你非常善于总结文本中的重要信息"),
        ("human","这是针对提问的完整解释说明内容：{description}"),
        ("human","请总结出问题的关键信息，20字以内"),
    ]
)
chainB_chains = LLMChain(
    llm=llm,
    prompt=chainB_template,
    verbose=True,
)
```

```python
from langchain_classic.chains.sequential import SimpleSequentialChain
full__chain = SimpleSequentialChain(
    chains=[chainA_chains, chainB_chains],
    verbose=True,
)
#SimpleSequentialChain的唯一输入变量名是：input
response=full__chain.invoke({"input":"如何使用LangChain进行链式调用"})
print(response)
```

## 举例2：新版runnable写法
### runnable前一个输出作为后一个输入... | ... | ...称为Runnable串

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)

chainA = (
    ChatPromptTemplate.from_messages([
        ("system", "你是一个AI助手"),
        ("human", "我的问题是{question}"),
    ])
    | llm
    | StrOutputParser()
)

chainB = (
    ChatPromptTemplate.from_messages([
        ("system", "你非常善于总结文本中的重要信息"),
        ("human", "这是针对提问的完整解释说明内容：{description}"),
        ("human", "请总结答案20字以内"),
    ])
    | llm
    | StrOutputParser()
)

result_a = chainA.invoke({"question": "大模型agent开发岗位需要什么技能？"})
result_b = chainB.invoke({"description": result_a})
print(result_a)
print(result_b)
```

### SequentialChain的用法

```python
from langchain_classic.chains.sequential import SequentialChain
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import StrOutputParser

llm = ChatOpenAI(
    api_key=os.getenv("DEEPSEEK_API_KEY"),
    base_url=os.getenv("DEEPSEEK_BASE_URL"),
    temperature=0,
    model="deepseek-chat",
)
#两个输入
schianA_template=ChatPromptTemplate.from_messages(
    [
        ("system","你是一个AI助手"),
        ("human","请你先解释一下:{knowledge}，并且{action}"),
    ]
)
scahinA_chians=LLMChain(
    llm=llm,
    prompt=schianA_template,
    verbose=True,
    output_key="result_A",
)
schianB_template=ChatPromptTemplate.from_messages(
    [
        ("system","你非常善于总结文本中的重要信息"),
        ("human","这是针对提问的完整解释说明内容：{result_A}"),
        ("human","请总结50字以内"),
    ]
)
schianB_chians=LLMChain(
    llm=llm,
    prompt=schianB_template,
    verbose=True,
    output_key="result_b",
)
#一定要声明两个变量：input_variables和output_variables
Sq_chian=SequentialChain(
    chains=[scahinA_chians, schianB_chians],
    input_variables=["knowledge", "action"],
    output_variables=["result_A", "result_b"],##控制输出a和b的结果都有
    verbose=True,
)
response=Sq_chian.invoke({"knowledge":"中国足球为什么那么菜", "action":"举例20字以内"})
print(response)
print(type(response))#<class 'dict'>
```
